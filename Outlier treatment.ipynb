```python
import pandas as pd
```

```python
df=pd.read_csv("kaggle_merge.csv")
```

```python
df.drop(columns='Unnamed: 0',inplace=True,axis=1)
```

```python
df.describe(percentiles=[.25,0.5,0.75,0.90,0.95,0.995])
```

```python
def outlier_capping(x):
    if ((len(x.unique()) > 10)):
        x = x.clip(upper=x.quantile(0.995))
    
    return(x)
```

```python
#outlier treatment
for col in df:
    if df[col].dtype != 'object':
        df2 = df.apply(lambda x: outlier_capping(x))
```
